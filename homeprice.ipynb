```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt



```

```python
df = pd.read_csv('/content/drive/MyDrive/homprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.scatter(df.AREA,df.PRICE,color='red',marker='+')
```

```python
new_df=df.drop('PRICE',axis='columns')
new_df
```

```python
PRICE=df.PRICE
PRICE
```

```python
reg=linear_model.LinearRegression()
reg.fit(new_df,PRICE)
```

```python
reg.predict([[3000]])
```

```python
reg.coef_
```
